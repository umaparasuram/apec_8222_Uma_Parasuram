**Mac-specific note:** Throughout this course, I will only have time to teach the Windows version of the approach, but in most cases, the Mac version is basically the same, so I leave it to you to figure out the difference if you so choose. Here, however, is one place where Mac is not the same as PC: The Console (PC) vs. the Terminal (Mac). To get to Bash on Mac, follow these instructions:

-   Open the mac terminal (search "terminal" under Spotlight Search line)
-   Call "bash" so that the terminal is operating in the bash language. Bash is an enhanced set of commands that let you interact with files and remote servers.
-   Set the working directory to your desired directory path. On the PC version, I took the shortcut of doing this by right-clicking on the appropriate directory. The mac version doesn’t have that so instead navigate to the correct directory with something like:
    -   `cd/Users/your_username/APEC_8222/Code`
-   Note that bash stops reading if it encounters a blank, so I had to change the name of a file from APEC 8222 to APEC_8222
-   Now the current working directory is set up, run git clone \[link\] The Mac-specific steps are summarized in this fig: ![](images/assignment1_git_image5.png){#fig-githw_5 fig-align="center" width="37%"}

Note a couple of key files in here. First, the transparent `.git` folder will not show up on your machine unless you have enabled the ability to see hidden files. Finally, the `.gitignore` text file has a list of files or directories you don’t want to show up in Git, Git Desktop, or other tools. Don’t worry about this right now.

# Step 4: Create your own repository on GitHub

We are going to create SEPARATE repositories for the course code and your code. This makes sense because even though you can clone the course repository, you cannot push changes to the server (unless I add you as a developer via GitHub). The repo you create, you will of course be able to push to.

1.  Go to github.com, login with your credentials, and click on the repository tab. From there, click the green button “Add”. This will open a window like below. Enter the name for your repository as “big_data” (making it exactly this will make it easier to use the commands I provide below).

![](images/assignment1_git_image8.png){#fig-githw_8 width="100%" fig-align="center"}

2.  Now, the repo exists online but not yet on your local machine. We are going to clone this to your machine now, similar to above. Note that it is possible to do this in the opposite order (e.g., create a local repository, connect it to a remote, then push it) but this is more likely to mess up if you do it in the wrong order. Conveniently, in the screen that shows up after you create your repo, you will find the git link you need. Copy this.

![](images/assignment1_git_image9.png){#fig-githw_9 width="100%" fig-align="center"}

3.  Just to be safe, close down your old Git Bash and open a NEW one in the Code directory. Enter the clone command like before, but now with your new link. It will warn you that the repo is empty but that’s okay.

![](images/assignment1_git_image10.png){#fig-githw_10 width="100%" fig-align="center"}

4.  At this point, you should now have two directories in your Code folder, one for the class contents and one that you will use for whatever you want, but also for pushing your Python assignments. Note that you should not change anything in the course repo (the python_for_big_data), though if you do it won’t mess anything up for others. Your folders should look like this:

# Step 5: Add a new file and push it

1.  Go to your repository, ‘big_data’, and add a new text file named ‘assignment_1.md’. You can do this however you want, though I think the easiest way is to right-click on the Window’s Explorer folder and under New select Text File. This will create “New Text Document.txt” which you can just rename assignment_1.md. Note that by default, Windows will not show file extensions, which means you can’t change the file extension. You can enable showing and editing file extensions by clicking View -\> Options in windows explorer and then unchecking “Hide extensions for known file types,” as shown in the fig below.

![](images/assignment1_git_image12.png){#fig-githw_12 width="100%" fig-align="center"}

2.  Edit the contents of assignment_1.md using a text editor to contain some new text (of whatever you want). You can use whatever text editor you want, but I’d suggest using Pycharm because that’s what you’ll use on the next assignment.
3.  Using a new Git Bash window in your big_data repo (close the prior one if you haven’t already), call the command ‘`git add assignment_1.md`’, as shown below. If you type git status”, also in below, you can see that there is a new file added, but it is listed under “`Changes to be committed`.”

![](images/assignment1_git_image13.png){#fig-githw_13 width="100%" fig-align="center"}

4.  Now git commit using “`git commit -m “Some message`”.” This shows up as:

![](images/assignment1_git_image14.png){#fig-githw_14 width="100%" fig-align="center"}

5.  Git Push this commit to your GitHub:

![](images/assignment1_git_image15.png){#fig-githw_15 width="100%" fig-align="center"}
